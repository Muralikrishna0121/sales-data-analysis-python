```python
df = pd.read_csv("data/sales_data.csv", encoding="latin1")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"])
```

```python
df["ORDERDATE"].head()
```

```python
total_sales = df["SALES"].sum()
avg_sales = df["SALES"].mean()
max_sales = df["SALES"].max()
min_sales = df["SALES"].min()

print("Total Sales:", total_sales)
print("Average Sales:", avg_sales)
print("Highest Sale:", max_sales)
print("Lowest Sale:", min_sales)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

df.groupby("PRODUCTLINE")["SALES"].sum().sort_values().plot(kind="barh")

plt.title("Sales by Product Line")
plt.xlabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(10,5))

df.groupby("COUNTRY")["SALES"].sum()\
.sort_values(ascending=False)\
.head(10)\
.plot(kind="bar")

plt.title("Top 10 Countries by Sales")
plt.ylabel("Sales")

plt.show()
```

```python
monthly_sales = df.groupby("MONTH_ID")["SALES"].sum()

plt.figure(figsize=(10,5))

monthly_sales.plot(marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df.groupby("YEAR_ID")["SALES"].sum().plot(kind="bar")

plt.title("Sales by Year")
plt.ylabel("Sales")

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))

sns.countplot(data=df, x="DEALSIZE")

plt.title("Deal Size Distribution")

plt.show()
```

```python
plt.figure(figsize=(12,5))

df.groupby("CUSTOMERNAME")["SALES"].sum()\
.nlargest(10)\
.plot(kind="bar")

plt.title("Top 10 Customers by Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x="STATUS")

plt.xticks(rotation=45)

plt.title("Order Status Distribution")

plt.show()
```
